### Apply preprocessing on xtrain and xtest

range of pixel values is 0-255

```python
xtrain_scl = xtrain/255
xtest_scl = xtest/255
```

```python
import numpy as np
np.unique(ytrain)
```

```python
len(np.unique(ytrain))
```

### Convert ytrain and ytest to_categorical

```python
from keras.utils import to_categorical
ytrain_cat = to_categorical(ytrain)
ytest_cat = to_categorical(ytest)
```

```python
ytrain_cat[0:5]
```

```python
ytest_cat[0:5]
```

### Create a convolution Neural Network

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPool2D, BatchNormalization, Dropout, Flatten, Dense
```

```python
model = Sequential()

# Create convolutional Layer 1 
model.add(Conv2D(filters=16, kernel_size=(3, 3), input_shape=(28, 28, 1), activation='relu'))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(BatchNormalization()) 
model.add(Dropout(0.1))

# Convolution Layer 2
model.add(Conv2D(filters=16, kernel_size=(3,3), activation='relu'))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(BatchNormalization())
model.add(Dropout(0.1))

#Flatten Layers
model.add(Flatten())

# Hidden Layer 
model.add(Dense(128, activation='relu'))
model.add(Dense(128, activation='relu'))

# Output Layer
model.add(Dense(10, activation='softmax'))
```

```python
model.summary()
```

### Compile the model

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'] )
```

### Train The model

```python
hist = model.fit(xtrain_scl, ytrain_cat, validation_split=0.2, epochs=5)
```

### Plot the learning curve

```python
plt.figure(figsize=(12, 6))
plt.plot(hist.history['loss'], label='Train Loss')
plt.plot(hist.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Categorical Crossentropy loss')
plt.title('Learning Curve')
plt.show()
```

### Evaluate model in train and test

```python
model.evaluate(xtrain_scl, ytrain_cat)
```

```python
model.evaluate(xtest_scl, ytest_cat)
```

### Predict Probabilities for train and test

```python
yprob_tr = model.predict(xtrain_scl)
yprob_ts = model.predict(xtest_scl)
```

```python
yprob_tr[0:5]
```

### Get Predictions

```python
ypred_tr = np.argmax(yprob_tr, axis=1)
ypred_ts = np.argmax(yprob_ts, axis=1)
```
