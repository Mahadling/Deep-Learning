# Image Processing

OpenCV Package

```python
import cv2
```

```python
img = cv2.imread('CuteCat.jpg')
img
```

### Displaying above image with matplotlib

```python
import matplotlib.pyplot as plt
plt.imshow(img)
plt.show()
```

### By defaultt cv2 reads image as BGR

### Converting BGR to RGB

```python
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img2)
plt.show()
```

### Splitting the RGB components of an image

```python
r, g, b = cv2.split(img2)
```

```python
r
```

```python
r.shape
```

```python
g
```

```python
b
```

```python
b.shape
```

### Visualizing RGB components

```python
plt.imshow(r, cmap='Reds')
plt.show()
```

```python
plt.imshow(g, cmap='Greens')
plt.show()
```

```python
plt.imshow(b, cmap='Blues')
plt.show()
```

### Read coloured image as grayscale

```python
img3 = cv2.imread('CuteCat.jpg', cv2.IMREAD_GRAYSCALE)
img3
```

```python
img3.shape
```

```python
plt.imshow(img3, cmap='gray')
plt.show()
```

### Reading a grayscale image

```python
img4 = cv2.imread('Image.jpg', cv2.IMREAD_GRAYSCALE)
img4
```

```python
plt.imshow(img4, cmap='gray')
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(12,12))
sns.heatmap(img4, annot=True, fmt='d')
plt.show()
```
